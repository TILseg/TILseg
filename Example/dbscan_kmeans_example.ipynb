#### 3. Inside def preprocess(), the def sort_patches function was edited to remove patches with too much white background. To understand how it works, we will extract the the dataframe of all the raw patches before white background removal.

* #### Code snippet from def main_preprocessing (def preprocess << def main_preprocessing) that generates a table of all the patches before background patches are removed --> `dataframe_patches`

```python
slide_path = repository_path + '/Example/Image_Files/TCGA-A2-A0CW-01Z-00-DX1.svs'
slide_file, slide_x, slide_y = open_slide(slide_path)

# calculate dimensions and losses
ydim, ypatch, yloss = get_tile_size(3000, slide_y)
xdim, xpatch, xloss = get_tile_size(4000, slide_x)

# get the dataframe for all patches in the slide
dataframe_patches = compile_patch_data(slide_file, ypatch, #Contains a column for average grey value of each patch
                                        xpatch, xdim, ydim)
```

* #### Let's visualize the distribution of average grey values of each patch from `dataframe_patches`

```python
grey_list = dataframe_patches['greys']
plt.hist(grey_list, bins=50, color='lightblue', edgecolor='black');
plt.xlabel('Grayscale Values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

* #### It is clear that there are two peaks - one at ~175 and another at ~255. Pure white images have a grey scale of 255 and images that contain a lot of white or low saturation colors will also be closer to this 255 peak. Since we want to eliminate the patches that contain too much white, we want to identify the means (or centers) of these two peaks and determine a greyscale cutoff to eliminate background patches (this will be based off the standard deviation from the ~175 peak). Let's do that below!

```python
#Create Guassian Distribution with 2 components
gmm = GaussianMixture(n_components=2)
grey_list =  np.array(grey_list).reshape(-1,1)
gmm.fit(grey_list)

# Get means and standard deviations of the two Gaussian components
means = gmm.means_.flatten()
std_devs = np.sqrt(gmm.covariances_).flatten()

# Sort means and std_devs to identify higher and lower peaks
sorted_indices = np.argsort(means)
lower_peak_mean, higher_peak_mean = means[sorted_indices]
lower_peak_std_dev, higher_peak_std_dev = std_devs[sorted_indices]

# Calculate cutoff
cutoff_value_lower_peak = lower_peak_mean + 0.25 * lower_peak_std_dev

#Plotting
plt.hist(grey_list, bins=50, color='lightblue', edgecolor='black');
plt.axvline(lower_peak_mean, color='red', linestyle='--', label=f'Peak 1 Mean: {round(lower_peak_mean,3)}')
plt.axvline(higher_peak_mean, color='green', linestyle='--', label=f'Peak 2 Mean: {round(higher_peak_mean,3)}')
plt.axvline(cutoff_value_lower_peak, color='red', linestyle='-', label=f'Peak 1 Cutoff: {round(cutoff_value_lower_peak,3)}')
plt.xlabel('Grayscale Values')
plt.ylabel('Frequency')
plt.title('Histogram with Peaks and Cutoffs')
plt.legend()
plt.grid(True)
plt.show()
```

#### The above histogram visualizes the peak identified at ~187 with a cutoff to remove background at ~192 based of 0.5 * standard deviation of the peak's mean. To understand what patches are being lost, run the following block to generate a plot showing images from each bin of gray values from 160 - 250:

```python
#Print min and max greys
print(f"min: {min(dataframe_patches['greys'])}")
print(f"max: {max(dataframe_patches['greys'])}")

# Define the ranges for the bins
bins = np.arange(160, 251, 10) #160 to 249 grayvalues with 10 values per bin

# Create a new column with the bin labels
dataframe_patches['grey_bin'] = pd.cut(dataframe_patches['greys'], bins=bins, labels=np.arange(len(bins)-1))

# Create a subplot grid
num_bins = len(bins) - 1
num_cols = 3  # Number of columns for subplots
num_rows = int(np.ceil(num_bins / num_cols))  # Calculate number of rows based on number of bins

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5*num_rows))

# Sample a random picture from each bin and plot in subplots (random state 13)
for i, ax in enumerate(axes.flat):
    if i < num_bins:
        # Select a random picture from the current bin
        sample = dataframe_patches[dataframe_patches['grey_bin'] == i].sample(1,random_state = 13)
        
        # Plot the picture
        ax.imshow(sample['patches'].values[0])  # Assuming 'image' is the column containing the pictures
        ax.set_title(f"Bin {i}: Greyscale Range {bins[i]}-{bins[i+1]-1}")
        ax.axis('off')
    else:
        # Hide empty subplots
        ax.axis('off')

plt.tight_layout()
plt.show()
```

* #### Visually, you can see that when transitioning between bin 3 (190 - 199) and bin 4 (200 - 209) the slides begin to have more background. This matches with the ~192 cutoff found from the Gaussian Distribution.
* #### Overall, this process was collapsed inside `def sort_patches`, which is called inside `def preprocess`, to remove the background

* #### It should be noted that not all .svs images may have a lot of background patches, meaning two peaks may not be necessary for finding the cutoff. This is why an if-statement inside `def sort_patches` checks for the coefficient of variance (coef_of_variance = standard deviation / mean), or COF
    * #### If the COF is less than 0.075 (arbitrary cuttoff unique to these slides), only a single peak's mean and standard deviation will be used to find the cutoff rather than the lower peak from a Guassian 2-component mixture.

### Case i: Running kmean_to_spatial_model_patch_wrapper on a Single Patch

#### 1. Run the patch version of the wrapper that takes in the filepath of the patch. Currently, there is no hyperarameter optimization for DBSCAN as it takes several hours to complete just a few parameters, so default values of `patial_hyperparameters = {'eps': 15,'min_samples': 100}` were used.
